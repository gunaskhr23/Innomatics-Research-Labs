Mobile Recharge Checker

```python

def recharge_process():
    plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))

            if amount < 50:
                print("Minimum recharge should be ₹50 or above.")
                continue

            if amount in plans:
                print(f"Recharge done successfully for ₹{amount}")
                break
            else:
                print("Selected plan is not available.")
                print("Available plans are:", plans)

        except:
            print("Invalid input. Please enter numbers only.")

recharge_process()
```

Stock Status Checker

```python

def stock_check(items):
    for item in items:
        qty = items[item]
        if qty < 15:
            print(f"{item} -> Need to Restock")
        else:
            print(f"{item} -> Sufficient Stock")

products = {
    "Laptop": 11,
    "Mouse": 25,
    "Keyboard": 18,
    "Monitor": 13
}

stock_check(products)
```

Student Marks Evaluation

```python

def result_status(marks):
    total = 0

    for m in marks:
        total += m

    avg = total / len(marks)
    print("Average:", avg)

    if avg >= 50:
        return "Student Passed"
    else:
        return "Student Failed"

marks_list = [65, 55, 72, 45]

final_result = result_status(marks_list)
print("Final Result:", final_result)
```

Cab Fare Calculator

```python

def fare_calculation(km, peak_time):
    base = 50
    rate = 12

    fare = base + (km * rate)

    if peak_time.lower() == "yes":
        fare = fare + (fare * 0.25)

    return fare

while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Peak hour? (yes/no): ")

    total = fare_calculation(distance, peak)
    print("Total Amount: ₹", total)

    again = input("Calculate again? (yes/no): ")
    if again.lower() != "yes":
        break
```

Attendance Percentage Checker

```python

def attendance_check(record):
    present = 0

    for day in record:
        if day == "P":
            present += 1

    percent = (present / len(record)) * 100
    print("Attendance:", percent, "%")

    if percent >= 75:
        return "Allowed for Exams"
    else:
        return "Not Allowed for Exams"

attendance_data = ["P", "A", "P", "P", "P", "A", "P", "P"]

status = attendance_check(attendance_data)
print("Status:", status)
```

Password Validator

```python

def password_check(pwd):
    special = "@#$"
    digit_found = False
    special_found = False

    if len(pwd) < 8:
        return "Password too short (Minimum 8 characters needed)"

    for ch in pwd:
        if ch.isdigit():
            digit_found = True
        if ch in special:
            special_found = True

    if digit_found and special_found:
        return "Password is Strong"
    else:
        return "Password must contain at least one number and one special symbol"

user_input = input("Enter password: ")
print(password_check(user_input))
```
